```python
# from config import api_key
import pandas as pd
import numpy as np
```

```python
# Data files
game_details_path = "games_details.csv"
games_path = "games.csv"
team_path = "teams.csv"

# Read the data files
game_details_data = pd.read_csv(game_details_path)
games_data = pd.read_csv(games_path)
teams_data = pd.read_csv(team_path)
```

```python
game_details_data.head()
```

```python
games_data.head()
```

```python
teams_data.head()
```

```python
# merging games details data with games data

merged = game_details_data.merge(games_data, on='GAME_ID')
merged
```

```python
# merging teams data to merged data

final_merged = merged.merge(teams_data, on='TEAM_ID')
final_merged
```

```python
final_merged.to_csv("output.csv", index=False)
```

```python
#only focus on season 2018
mask = final_merged.SEASON == 2018
mask1 = final_merged[mask]
mask1
```

```python
# excluding pre-season games
new_data = mask1[mask1["GAME_ID"] > 20000000]

new_data
```

# Stadium - Home Team wins

```python
stadium_DF = new_data[['GAME_ID','ARENA', 'TEAM_CITY', 'HOME_TEAM_WINS']].copy()
stadium_DF
```

```python
# Checking for number of stadiums
stadium_DF["ARENA"].nunique()
```

```python
# finding the total wins for each stadium
stadium_total = stadium_DF.groupby(["ARENA"]).nunique()["HOME_TEAM_WINS"].sum()
stadium_total

gameid_total = stadium_DF.groupby("ARENA")["GAME_ID"].nunique()
gameid_total


#create dataframe
stadium_sum = pd.DataFrame({
    "Total Wins": stadium_total,
    "Number of Games": gameid_total,
})
stadium_sum
```
